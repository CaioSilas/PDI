<p> 12) Concatenar a seguinte mariz três vezes de forma:
<ul>
<li>Horizontalmente</li>
<li>Verticalmente</li>
<li>Profundidade</li>
</ul>
</p>

````python
a = np.random.randint(1, 10, (3, 3))
b = np.random.randint(1, 10, (3, 3))
c = np.random.randint(1, 10, (3, 3))

# Concatenação horizontal
horizontal = np.concatenate((a, a, a), axis=1)
print("Horizontal:")
print(horizontal)
print(horizontal.shape)

# Concatenação vertical
vertical = np.concatenate((a, b, c), axis=0)
print("\nVertical:")
print(vertical)
print(vertical.shape)

# Concatenação em profundidade
depth = np.dstack((a, b, c))
print("\nProfundidade:")
print(depth)
print(depth.shape)
````

# Questões sobre Imagens

````python
def show(img_list, caption_list, Axis=False):
    n = len(img_list)
    f, ax = plt.subplots(1,n, figsize=(10+5*(n-1),10))
    for i in range(n):
        ax[i].imshow(img_list[i], cmap='gray')
        ax[i].set_title(caption_list[i])
        if not Axis:
            ax[i].set_axis_off()
````

<p>13) Trocar entre si os canais vermelho e azul. Ou seja, a imagem R,G,B vai ficar B,G,R</p>

````python
img = data.chelsea()
R,G,B = img[:,:,0], img[:,:,1], img[:,:,2]
nimg = np.dstack((B,G,R))
show([img, nimg], ['Original', 'Trocar'])
````

<p> 14) Rotacionar uma imagem em 90 graus sem utilizar funções que realizem a rotação de forma explicita</p>

````python
def rotate90(img):
  height, width, channels = img.shape
  nimg = np.zeros((width, height, channels), dtype=img.dtype)

  for i in range(height):
    for j in range(width):
        nimg[j, height - 1 - i] = img[i, j]

  return nimg
````

````python
img = data.chelsea()
nimg = rotate90(img)
show([img, nimg], ['0 graus', '90 graus'])
````
