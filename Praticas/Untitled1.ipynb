```python
!pip install scikit-fuzzy

import numpy as np
import skfuzzy as fuzz

# Gerando dados de exemplo
np.random.seed(42)
# Gere 100 pontos em torno de [1, 1] para o cluster 1
data1 = np.random.randn(100, 2) + np.array([1, 1])
# Gere 100 pontos em torno de [5, 5] para o cluster 2
data2 = np.random.randn(100, 2) + np.array([5, 5])
# Combine os dois conjuntos de dados
data = np.vstack([data1, data2])

# Aplicando o algoritmo FCM
cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
    data.T, 2, 2, error=0.005, maxiter=1000, init=None)

# Identificando a pertinência dos pontos a cada cluster
cluster_membership = np.argmax(u, axis=0)

# Mostrando os resultados
print("Centróides:")
print(cntr)
print("Pertinência dos primeiros 10 pontos:")
print(u[:,:10])
print("Pertinência dos clusters dos primeiros 10 pontos:")
print(cluster_membership[:10])

```
